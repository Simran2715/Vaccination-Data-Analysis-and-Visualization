```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(data=coverage_data, x='year',y='coverage')
plt.xlabel('Year')
plt.ylabel('Coverage')
```

```python
sns.pairplot(vaccine_schedule,hue='who_region')
```

```python
numeric_df = coverage_data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(coverage_data['target_number'], kde=True, bins=30)
plt.title('Distribution of Doses')
plt.xlabel('Doses')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.barplot(data=vaccine_schedule, x='year', y='schedule_rounds', hue='who_region')
plt.show()
```

```python
intro = vaccine_introduction['intro'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(intro,labels= intro.index,autopct='%1.1f%%', startangle=90)
plt.title('Vaccine Introduction')
plt.axis('equal')
plt.show()
```
